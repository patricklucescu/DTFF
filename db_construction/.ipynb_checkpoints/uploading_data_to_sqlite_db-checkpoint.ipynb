# uploading_data_to_sqlite_db

```python
import os
import sys
from pathlib import Path
import pandas as pd
import sqlalchemy as db
import sqlite3
```

Append path directory to data

```python
#setting paths at runtime in jupyter notebook does not work - we'll need a variable therefore 
path = os.path.join(Path(os.getcwd()).parent, 'data', '') 
```

### Database settings

```python
# SQLite
database_path = os.path.join(path, 'dtff_database.sqlite3')
engine = db.create_engine('sqlite:///' + database_path)
sqlite_db_con = sqlite3.connect(database_path)
```

### classification_risk_or_no_risk

```python
classification_risk_or_no_risk = pd.read_csv(path + 'classification_risk_or_no_risk.csv')
classification_risk_or_no_risk
```

```python
# todo: define dataytypes - common approach not working with sqlite
# classification_risk_or_no_risk_dtypes = {
#     'high_risk_environment': db.sql.sqltypes.INT()
# }
```

```python
classification_risk_or_no_risk.to_sql('classification_risk_or_no_risk', 
                                      con=sqlite_db_con, 
                                      index=False,
                                      if_exists='replace')
```

### global_news_sentiment_macro

```python
global_news_sentiment_macro = pd.read_csv(path + 'global_news_sentiment_macro.csv')
global_news_sentiment_macro
```

```python
global_news_sentiment_macro.to_sql('global_news_sentiment_macro', 
                                  con=sqlite_db_con, 
                                  index=False,
                                  if_exists='replace')
```

### returns

```python
returns = pd.read_csv(path + 'returns.csv')
returns
```

```python
returns.to_sql('returns', 
              con=sqlite_db_con, 
              index=False,
              if_exists='replace')
```
